```python
def dec_to_binary_signed(num_input, precision=10):
    """
    Convert signed integer/float string or number to a string
    of the form: '<orig> in binary number system is S|IIII.FFFFF...'
    where S is sign bit (0 positive, 1 negative).
    precision = number of fractional bits.
    """
    # accept either string or numeric
    try:
        num = float(num_input)
    except Exception:
        raise ValueError("Invalid input. Provide an integer or float (e.g. 32 or -10.25).")

    sign_bit = "0" if num >= 0 else "1"
    num_abs = abs(num)

    integer_part = int(num_abs)
    fractional_part = num_abs - integer_part

    # integer to binary
    integer_binary = bin(integer_part)[2:]  # '0b' removed

    # fractional to binary by repeated multiplication
    frac_bits = []
    temp = fractional_part
    for _ in range(precision):
        temp *= 2
        bit = int(temp)
        frac_bits.append(str(bit))
        temp -= bit
        if temp == 0:
            break

    if fractional_part == 0:
        binary_repr = f"{sign_bit}|{integer_binary}"
    else:
        frac_str = ''.join(frac_bits).ljust(precision, '0')
        binary_repr = f"{sign_bit}|{integer_binary}.{frac_str}"

    return f"{num_input} in binary number system is {binary_repr}"
```
